```python


# Initialize the WebDriver (use the appropriate driver for your browser, e.g., chromedriver)
driver = webdriver.Chrome()  # You can replace this with Firefox(), Edge(), etc.

try:
    # Open the desired webpage
    url = f"https://www.coursera.org/search?query=free&topic=Computer%20Science&sortBy=BEST_MATCH"
    driver.get(url)  # Replace with the URL of your target page
    
    # Wait for the page to load (adjust the time as needed based on the page's complexity)
    #time.sleep(3)

    # Get page height and scroll incrementally to load all content
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    while True:
        # Scroll down by simulating the END key press
        driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
        
        # Wait for new content to load (adjust the time as needed)
        time.sleep(10)
        
        # Check if the page height has stopped changing (end of the page)
        new_height = driver.execute_script("return document.body.scrollHeight")
        time.sleep(10)
        if new_height == last_height:
            break
        last_height = new_height

    # Get the full page source after scrolling
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # extrace prices and ars names from page
    institution = soup.find_all('p', {'class':"cds-ProductCard-partnerNames css-vac8rf"})
    rate  = soup.find_all('span', {'class':"css-6ecy9b"})
    title  = soup.find_all('h3', {'class':"cds-CommonCard-title css-6ecy9b"})
    skills  = soup.find_all('div', {'class':"cds-CommonCard-bodyContent"})
    reviews  = soup.find_all('div', {'class':"css-vac8rf"})
    level  = soup.find_all('div', {'class':"cds-CommonCard-metadata"})
    # clean them
    clean_rate = []
    clean_Institution = []
    clean_title = []
    clean_skills = []
    clean_reviews = []
    clean_level = []

    for i,p,l,r,s,t in zip(institution,rate,level,reviews,skills,title):
        clean_Institution.append(i.text)
        clean_rate.append(p.text)
        clean_title.append(t.text)
        clean_skills.append(s.text)
        clean_reviews.append(r.text)
        clean_level.append(l.text)

    # store them 
    new_rows_df = pd.DataFrame(
        {
            'institution': clean_Institution,
            'rate' : clean_rate ,
            'title':clean_title,
            'reviews':clean_reviews,
            'skills':clean_skills,
            'level':clean_level,
        })

    df_cs = pd.concat([df_cs, new_rows_df], ignore_index=True)

    
finally:
    # Close the browser
    driver.quit()
```

```python
columns_ = ['institution','title','rate','reviews','skills','level']
df_ds = pd.DataFrame(columns=columns_)
```

```python
# Initialize WebDriver
driver = webdriver.Chrome()

# Open the webpage
url = "https://www.coursera.org/search?query=free&topic=Data%20Science&sortBy=BEST_MATCH"
driver.get(url)

try:
    # Wait for the page to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "body")))
    
    # Start scrolling
    last_height = driver.execute_script("return document.body.scrollHeight")
    scroll_attempts = 0  # Counter to avoid infinite loops

    while scroll_attempts < 10:  # Limit attempts to avoid infinite loop
        # Scroll down by a small step
        driver.execute_script("window.scrollBy(0, 1500);")
        time.sleep(2)  # Wait to allow new content to load
        
        # Get the new height
        new_height = driver.execute_script("return document.body.scrollHeight")
        
        if new_height == last_height:
            # If no new content is loaded, increase counter and wait again
            scroll_attempts += 1
            time.sleep(2)  # Wait before trying again
        else:
            # Reset counter if new content is loaded
            scroll_attempts = 0
            last_height = new_height
    
    # Parse the final page content
    soup = BeautifulSoup(driver.page_source, "html.parser")
    
    # Extract data as before
    institution = soup.find_all('p', {'class': "cds-ProductCard-partnerNames css-vac8rf"})
    rate = soup.find_all('span', {'class': "css-6ecy9b"})
    title = soup.find_all('h3', {'class': "cds-CommonCard-title css-6ecy9b"})
    skills = soup.find_all('div', {'class': "cds-CommonCard-bodyContent"})
    reviews = soup.find_all('div', {'class': "css-vac8rf"})
    level = soup.find_all('div', {'class': "cds-CommonCard-metadata"})
    
    # Clean and store data
    data = []
    for i, p, l, r, s, t in zip(institution, rate, level, reviews, skills, title):
        data.append({
            'institution': i.text.strip(),
            'rate': p.text.strip(),
            'title': t.text.strip(),
            'reviews': r.text.strip(),
            'skills': s.text.strip(),
            'level': l.text.strip(),
        })
    
    # Convert to DataFrame
    df_ds = pd.DataFrame(data)
    

except Exception as e:
    print(f"An error occurred: {e}")
finally:
    driver.quit()
```

```python
df_ds.to_csv('Data Science.csv', index=False)
```
